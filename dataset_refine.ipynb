```python
import pandas as pd
import os

xlsx_files = [f for f in os.listdir('./공장등록통계') if f.endswith('.xlsx')]

result_df = pd.DataFrame(columns=['지역명', '연', '월', '등록공장수'])

for file in xlsx_files:
    filename = os.path.join('.', file)
    year = int(filename.split('.')[1][2:6])  # Extract year from filename
    month = int(filename.split('.')[2][0:2])  # Extract month from filename
    print(f"Processing file: {filename}")
    print(f"Year: {year}, Month: {month}")

    df = pd.read_excel('./공장등록통계/' + filename, sheet_name='지역별 등록현황')
    # B3부터 B20까지의 데이터를 지역명 순서대로 추가
    new_df = pd.DataFrame({
        '지역명': df.iloc[2:20, 0].values,
        '연': year,
        '월': month,
        '등록공장수': df.iloc[2:20, 1].values,
    })
    result_df = pd.concat([result_df, new_df], ignore_index=True)


# sort result_df by 지역명, 연, 월
result_df.sort_values(by=['지역명', '연', '월'], inplace=True)
result_df.dropna(inplace=True)
# 지역명이 합계면 drop
result_df = result_df[result_df['지역명'] != '합계']
# 강원특별자치도를 '강원도'로 변경하고 등록공장수를 같은 연 월의 강원도와 합산
result_df.loc[result_df['지역명'] == '강원특별자치도', '지역명'] = '강원도'
result_df = result_df.groupby(['지역명', '연', '월'], as_index=False).sum()
# 전북특별자치도를 '전라북도'로 변경하고 등록공장수를 같은 연 월의 전라북도와 합산
result_df.loc[result_df['지역명'] == '전북특별자치도', '지역명'] = '전라북도'
result_df = result_df.groupby(['지역명', '연', '월'], as_index=False).sum()

result_df.to_csv('공장등록수.csv', index=False, encoding='utf-8')
# 지역명뜰 표시
result_df.groupby('지역명').size()
```

```python
import pandas as pd

df = pd.read_csv('202001_202407_주민등록인구및세대현황_월간.csv')

result_df = pd.DataFrame(columns=['지역명', '연', '월', '총인구수', '세대수', '세대당 인구', '남자 인구수', '여자 인구수'])

# "2020년01월_총인구수","2020년01월_세대수","2020년01월_세대당 인구","2020년01월_남자 인구수","2020년01월_여자 인구수"
# 해당 식 처럼 202001부터 202407까지의 컬럼을 연: 2021, 월: 1로 변환
# Filter only columns that contain '총인구수'
population_cols = [col for col in df.columns[1:] if '총인구수' in col]

def remove_non_numeric(value):
    """Remove non-numeric characters from a string."""
    return ''.join(filter(str.isdigit, str(value)))

for col in population_cols:
    year = int(col[:4])
    month = int(col[5:7])
    
    # Create column names for other statistics
    base_name = col.replace('총인구수', '')
    세대수_col = base_name + '세대수'
    세대당인구_col = base_name + '세대당 인구'
    남자인구_col = base_name + '남자 인구수'
    여자인구_col = base_name + '여자 인구수'
    
    new_df = pd.DataFrame({
        '지역명': df['행정구역'],
        '연': year,
        '월': month,
        '총인구수': df[col].apply(lambda x: int(remove_non_numeric(x)) if pd.notnull(x) else 0),
        '세대수': df[세대수_col].apply(lambda x: int(remove_non_numeric(x)) if pd.notnull(x) else 0) if 세대수_col in df.columns else None,
        '세대당 인구': df[세대당인구_col].apply(lambda x: float((x)) if pd.notnull(x) else 0) if 세대당인구_col in df.columns else None,
        '남자 인구수': df[남자인구_col].apply(lambda x: int(remove_non_numeric(x)) if pd.notnull(x) else 0) if 남자인구_col in df.columns else None,
        '여자 인구수': df[여자인구_col].apply(lambda x: int(remove_non_numeric(x)) if pd.notnull(x) else 0) if 여자인구_col in df.columns else None,
    })
    result_df = pd.concat([result_df, new_df], ignore_index=True)
# sort result_df by 지역명, 연, 월
result_df.sort_values(by=['지역명', '연', '월'], inplace=True)
result_df.dropna(inplace=True)
result_df = result_df[result_df['지역명'] != '전국']
# 강원특별자치도를 '강원도'로 변경하고 등록공장수를 같은 연 월의 강원도와 합산
result_df.loc[result_df['지역명'] == '강원특별자치도', '지역명'] = '강원도'
result_df = result_df.groupby(['지역명', '연', '월'], as_index=False).sum()
# 전북특별자치도를 '전라북도'로 변경하고 등록공장수를 같은 연 월의 전라북도와 합산
result_df.loc[result_df['지역명'] == '전북특별자치도', '지역명'] = '전라북도'
result_df = result_df.groupby(['지역명', '연', '월'], as_index=False).sum()

result_df.to_csv('주민등록인구및세대현황.csv', index=False, encoding='utf-8')
# 지역명뜰 표시
# result_df.groupby('지역명').size()
result_df
```

```python
import pandas as pd
import os

csvs = [f for f in os.listdir('./기온') if f.endswith('.csv')]

result_df = pd.DataFrame(columns=['지역명', '연', '월', '평균기온', '평균최저기온', '평균최고기온'])

for csv in csvs:
    filename = csv
    print(f"Processing file: {filename}")

    df = pd.read_csv('./기온/' + filename)
    
    # '지역명' 컬럼을 '지점명'으로 변경
    df.rename(columns={'지점명': '지역명'}, inplace=True)
    
    new_df = pd.DataFrame({
        '지역명': csv.split('.')[0],  # Use the filename without extension as 지역명
        '연': df['년월'].apply(lambda x: int(str(x)[:4])),
        '월': df['년월'].apply(lambda x: int(str(x)[5:7])),
        '평균기온': df['평균기온(℃)'],
        '평균최저기온': df['평균최저기온(℃)'],
        '평균최고기온': df['평균최고기온(℃)'],
    })
    
    result_df = pd.concat([result_df, new_df], ignore_index=True)  

# sort result_df by 지역명, 연, 월
result_df.sort_values(by=['지역명', '연', '월'], inplace=True)
result_df.dropna(inplace=True)

result_df.to_csv('기온.csv', index=False, encoding='utf-8')
result_df
```

```python
import pandas as pd
import os

json_files = [f for f in os.listdir('./전기') if f.endswith('.json')]

result_df = pd.DataFrame(columns=['지역명', '연', '월', '대상가구수', '평균사용량', '평균금액'])
for file in json_files:
    filename = os.path.join('./전기', file)
    print(f"Processing file: {filename}")

    df = pd.read_json(filename, encoding='utf-8')
    
    new_df = pd.DataFrame({
        '지역명': file.split('.')[0],  # Use the filename without extension as 지역명
        '연': df['연'].apply(lambda x: int(x)),
        '월': df['월'].apply(lambda x: int(x)),
        '대상가구수': df['대상가구수'],
        '평균사용량': df['평균사용량'],
        '평균금액': df['평균금액'],
    })
    
    result_df = pd.concat([result_df, new_df], ignore_index=True)
  
# sort result_df by 지역명, 연, 월
result_df.sort_values(by=['지역명', '연', '월'], inplace=True)
result_df.dropna(inplace=True)
result_df.to_csv('전국평균전력사용량.csv', index=False, encoding='utf-8')
result_df
```
